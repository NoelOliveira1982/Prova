## 6.1 — Reflexão sobre o Pipeline de Aprendizado

O **tratamento de valores ausentes** é a primeira etapa crítica: colunas com percentual extremo de ausência, como `weight` (97%), foram removidas para evitar imputações ruidosas que distorceriam as relações entre features e target; já para `race` (2,2%), a imputação pela moda preservou a distribuição original sem introduzir viés significativo. Sem esse tratamento, algoritmos como Árvore de Decisão e MLP falhariam ou produziriam modelos enviesados.

A **codificação de variáveis categóricas** transforma atributos textuais em representações numéricas compatíveis com os algoritmos. Utilizamos Ordinal Encoding para variáveis com ordem natural (ex.: faixas etárias) e One-Hot Encoding para nominais (ex.: raça, gênero), evitando que o modelo interprete relações ordinais inexistentes. Contudo, o One-Hot pode gerar alta dimensionalidade e esparsidade, impactando o cálculo de distâncias no K-Means e a convergência da MLP.

A **seleção de atributos** por importância (Random Forest) reduziu de 50+ features para as Top-30 mais relevantes, eliminando ruído e redundância. Isso diminuiu o overfitting, acelerou o treinamento e manteve desempenho comparável (F1-macro), além de produzir clusters mais coesos na etapa não supervisionada. Além disso, o uso de **StratifiedGroupKFold** (agrupando por paciente) na validação cruzada foi essencial para evitar data leakage: como 29,7% dos registros são de pacientes com múltiplos encontros, uma divisão ingênua permitiria que o modelo "visse" dados do mesmo paciente no treino e no teste, inflando artificialmente as métricas. Em conjunto, essas etapas formam um pipeline interdependente: cada decisão de pré-processamento propaga seus efeitos ao longo de toda a cadeia de modelagem.

## 6.2 — Conclusão

A reflexão acima demonstra que o pré-processamento não é uma etapa mecânica, mas um conjunto de decisões estratégicas que impactam diretamente a qualidade dos modelos supervisionados e não supervisionados. As escolhas realizadas neste projeto — desde a remoção de colunas com alta ausência até a seleção por importância via Random Forest — foram fundamentadas nos dados e refletidas nos resultados obtidos nas Questões 2 a 5.
